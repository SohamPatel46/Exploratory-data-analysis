```python
import seaborn as sb
```

```python
s=sb.load_dataset("tips")   #inbuilt data 
s.head()
```

```python
s.dtypes
```

```python
#Correlation with Heatmap(features)
#It helps to find interrelation between features (only of continous value)
```

```python
s.corr()
```

```python
sb.heatmap(s.corr())
```

```python
#Joint plot (use for bivariant)(relation betwen 2 variable)
sb.jointplot(x='tip',y='total_bill',data=s,kind="reg")  #hex/reg
```

```python
#pair plot ( more than 2 features)
sb.pairplot(s)
```

```python
sb.pairplot(s,hue='sex') #sex/smooker
```

```python
#Dist plot(similar to histogram)(Y axis will have pdf)

sb.distplot(s["tip"])
```

```python
#Catagorical Plots 
```

```python
#Boxplot
sb.boxplot('day','tip',hue='sex',data=s)  #remove hue if dont want to partition male and female
```

```python
#countplot (count of x in y)
sb.countplot('day',data=s)    #sex/smoker/time
```

```python
#bar plot (x in respect to y)
sb.barplot(x="tip",y='sex',data=s)
```
